Vi ser på et enkelt skjema for å modellere et strikkhopp.

## Modellen

Vi starter med Newtons 2e lov:

<!-- Equation labels as ordinary links -->
<div id="_auto1"></div>

$$
\begin{equation}
\sum F = m\cdot a = m\cdot \frac{dv}{dt}
\label{_auto1} \tag{1}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto2"></div>

$$
\begin{equation} 
\frac{dv}{dt} = \frac{\sum F}{m}
\label{_auto2} \tag{2}
\end{equation}
$$

Summen av kreftene $\sum F$ er i dette tilfellet lik gravitasjonen, $G=m\cdot g$ + strikkraften, $S=-k\cdot v$ og luftmotstandet $L=-D\cdot v$

## Implementasjon

```python
# Inline plotting inne i jupyter!
%matplotlib inline

# Importerer numeriske metoder vi trenger
from pylab import *

# Parametrar
v0 = 0.0    # Utgangsfart
x0 = 20     # Utgangshøyde
            
m  = 70.    # Masse
g  = 9.81   # Tyngdens akselerasjon
k  = 40     # Strikkonstanten
D  = 10     # Luftmotstandskonstant
xL = 30     # Strikk lengde

dt = 1e-3   # Tidssteg
tstop = 60  # Hvor lenge skall simuleringen vare?
N = int(ceil(T/dt))	# Lengde på liste 

# Initialiserer lister som skall ha løsningene
v = zeros(N+1)
x = zeros(N+1)
t = linspace(0, T, N+1)

# Setter første elementet i listene til utgangsfarten og utgangshøyden 
v[0] = v0
x[0] = x0

# Definerer strikkraften (ikke noen kraft hvis strikken er slak)
def S(x):
    if x0 - x < xL:
        print "bada", x
        return 0
    else:
        print "jada", x
        return -k*x

def a(x, v):
    return S(x)/m - g - D/m*v

for i in range(N):
    v[i+1] = v[i] + a(x[i], v[i])*dt
    x[i+1] = x[i] + v[i]*dt + 0.5*a(x[i], v[i])*dt**2

plot(t, x)
grid()
show()
```
